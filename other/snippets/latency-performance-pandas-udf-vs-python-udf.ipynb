```python
import random
import string
import time
import pandas as pd

# Generate sample data: 250 rows x 20 columns (mixed ints and strings)
def generate_sample_data(rows=250, cols=20):
    data = []
    for _ in range(rows):
        row = []
        for c in range(cols):
            if c % 2 == 0:
                row.append(random.randint(1, 1000))
            else:
                row.append(''.join(random.choices(string.ascii_letters, k=5)))
        data.append(row)
    return data

# Process rows using array of arrays (pure Python)
def process_rows_array(rows, sort_column_index):
    sorted_rows = sorted(rows, key=lambda x: x[sort_column_index], reverse=True)
    return sorted_rows[:5]

# Process rows using Pandas DataFrame
def process_rows_pandas(rows, column_names, sort_column_name):
    df = pd.DataFrame(rows, columns=column_names)
    return df.sort_values(sort_column_name, ascending=False).head(5)

def process_rows_pandas_no_create(df, sort_column_name):
    return df.sort_values(sort_column_name, ascending=False).head(5)


def main():
    rows, cols = 250, 20
    data = generate_sample_data(rows, cols)

    sort_col_index = 0
    column_names = [f"col_{i}" for i in range(cols)]
    sort_col_name = column_names[sort_col_index]

    # Timing array-of-arrays method
    start = time.perf_counter()
    top5_array = process_rows_array(data, sort_col_index)
    end = time.perf_counter()
    print("Top 5 rows (Array of Arrays):", top5_array, sep="\n")
    print(f"Array of Arrays processing time: {(end - start)*1000:.3f} ms\n")

    # Timing Pandas DataFrame method
    start = time.perf_counter()
    top5_pandas = process_rows_pandas(data, column_names, sort_col_name)
    end = time.perf_counter()
    print("Top 5 rows (Pandas DataFrame):", top5_pandas, sep="\n")
    print(f"Pandas DataFrame processing time: {(end - start)*1000:.5f} ms")

    # Timing Pandas DataFrame method - precreate DataFrame
    df = pd.DataFrame(data, columns=column_names)
    start = time.perf_counter()
    top5_pandas = process_rows_pandas_no_create(df, sort_col_name)
    end = time.perf_counter()
    print("Top 5 rows (Pandas DataFrame):", top5_pandas, sep="\n")
    print(f"Pandas DataFrame (no create DataFrame) processing time: {(end - start)*1000:.5f} ms")


if __name__ == "__main__":
    main()
```

```python
import numpy as np

#    A function that can be executed either as a Python function or as a Pandas UDF.
#    It takes 4 numerical parameters and returns the maximum value.
def max_param(param1, param2, param3, param4):
    start = time.perf_counter()
    result = np.maximum.reduce([param1, param2, param3, param4])
    end = time.perf_counter()
    print(f"Execution Time: {(end - start) * 1000:.6f} milliseconds")    
    return result

# Example usage as a regular Python function
result_python = max_param(1.0, 2.0, 3.0, 4.0)

# Example usage as a Pandas UDF
batch_size = 2500000
data = pd.DataFrame(np.random.rand(batch_size, 4), columns=['param1', 'param2', 'param3', 'param4'])

# Execute as a Pandas UDF on a batch of rows
results_batch = max_param(data['param1'], data['param2'], data['param3'], data['param4'])
```

```python
import time
import numpy as np
import pandas as pd

def max_param(param1, param2, param3, param4):
    """
    A function that can be executed in Python mode.
    It takes 4 numerical parameters and returns the maximum value.
    """
    start = time.perf_counter()
    result = np.maximum.reduce([param1, param2, param3, param4])
    end = time.perf_counter()
    # print(f"Execution Time: {(end - start) * 1000:.6f} milliseconds")    
    return result

# Example usage as a regular Python function
result_python = max_param(1.0, 2.0, 3.0, 4.0)
print("Result (Python mode):", result_python)

# Generate a batch of 2.5 million rows with random values
batch_size = 2500000
data = np.random.rand(batch_size, 4)

# Execute the function in Python mode on each row
start_batch = time.perf_counter()
results_batch = [max_param(row[0], row[1], row[2], row[3]) for row in data]
end_batch = time.perf_counter()

print(f"Batch Execution Time: {(end_batch - start_batch) * 1000:.3f} milliseconds")

# Display first 10 results
print("First 10 Results:")
print(results_batch[:10])
```
