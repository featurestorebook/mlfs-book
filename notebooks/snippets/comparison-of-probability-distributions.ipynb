```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, lognorm

# Parameters for the distributions
lambda_exp = 1  # Rate parameter for exponential distribution
mu_ln = 0  # Mean of log(X) for log-normal distribution
sigma_ln = 0.5  # Standard deviation of log(X) for log-normal distribution

# Generate x values
x = np.linspace(0.01, 5, 1000)

# Calculate the PDF for exponential distribution
pdf_expon = expon.pdf(x, scale=1/lambda_exp)

# Calculate the PDF for log-normal distribution
pdf_lognorm = lognorm.pdf(x, s=sigma_ln, scale=np.exp(mu_ln))

# Plot the distributions
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_expon, label='Exponential Distribution', color='blue')
plt.plot(x, pdf_lognorm, label='Log-Normal Distribution', color='red')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Comparison of Exponential and Log-Normal Distributions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import binom, poisson

# Parameters for the distributions
n_binom = 20  # Number of trials for binomial distribution
p_binom = 0.5  # Probability of success for binomial distribution
lambda_poisson = n_binom * p_binom  # Mean for Poisson distribution (same as np for binomial)

# Generate x values
x = np.arange(0, 30)

# Calculate the PMF for binomial distribution
pmf_binom = binom.pmf(x, n_binom, p_binom)

# Calculate the PMF for Poisson distribution
pmf_poisson = poisson.pmf(x, lambda_poisson)

# Plot the distributions
plt.figure(figsize=(10, 6))
plt.plot(x, pmf_binom, marker='o', linestyle='-', color='blue', label='Binomial Distribution')
plt.plot(x, pmf_poisson, marker='o', linestyle='-', color='red', label='Poisson Distribution')
plt.xlabel('x')
plt.ylabel('Probability Mass')
plt.title('Comparison of Binomial and Poisson Distributions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import norm, uniform

# Parameters for the distributions
mu_norm = 0  # Mean for normal distribution
sigma_norm = 1  # Standard deviation for normal distribution
a_uniform = -3  # Lower bound for uniform distribution
b_uniform = 3   # Upper bound for uniform distribution

# Generate x values
x = np.linspace(-5, 5, 1000)

# Calculate the PDF for normal distribution
pdf_norm = norm.pdf(x, mu_norm, sigma_norm)

# Calculate the PDF for uniform distribution
pdf_uniform = uniform.pdf(x, loc=a_uniform, scale=b_uniform-a_uniform)

# Plot the distributions
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_norm, marker='o', linestyle='-', color='blue', label='Normal Distribution')
plt.plot(x, pdf_uniform, marker='o', linestyle='-', color='red', label='Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Comparison of Normal and Uniform Distributions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import norm, lognorm

# Parameters for the distributions
# Bimodal distribution parameters
mu1, sigma1 = -2, 0.5  # Mean and standard deviation for the first normal distribution
mu2, sigma2 = 3, 1  # Mean and standard deviation for the second normal distribution

# Skewed distribution parameters
mu_skew, sigma_skew = 0, 0.5  # Parameters for the log-normal distribution

# Generate x values
x = np.linspace(-5, 10, 1000)

# Calculate the PDF for the bimodal distribution
pdf_bimodal = 0.5 * norm.pdf(x, mu1, sigma1) + 0.5 * norm.pdf(x, mu2, sigma2)

# Calculate the PDF for the skewed (log-normal) distribution
pdf_skewed = lognorm.pdf(x, s=sigma_skew, scale=np.exp(mu_skew))

# Plot the distributions
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_bimodal, marker='o', linestyle='-', color='blue', label='Bimodal Distribution')
plt.plot(x, pdf_skewed, marker='o', linestyle='-', color='red', label='Skewed Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Comparison of Bimodal and Skewed Distributions')
plt.legend()
plt.grid(True)
plt.show()
```
